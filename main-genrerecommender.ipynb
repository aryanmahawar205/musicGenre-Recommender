```python
# importing all required modules and libraries

import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset
# in this dataset, gender 1 means Male and gender 0 means Female

musicData = pd.read_csv('/kaggle/input/music-recommendation/music.csv')
print(musicData)
```

```python
# splitting the dataset into two sets
# the first set will have age and gender columns which are essentially inputs to the model
# the second set will be the genre column which we expect as output from the model

X = musicData.drop(columns = ['genre']) #where X is the input set
y = musicData['genre'] #where y is the output set

# now splitting the dataset to further training and testing sets
# the testing set size is 20% of the total dataset, therefore training size is 80% of the total
# the train_test_split() function return a tuple which have been unpacked in the next line

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
# creating and training the model using Decision Tree Algorithms

model = DecisionTreeClassifier()
model.fit(X_train, y_train) #where X_train is the input training set and y_train is the output training set
```

```python
# making predictions from the model with accuracy score (0 - 1)

predictions = model.predict(X_test)
print(predictions)
accuracy = accuracy_score(y_test, predictions)
print(accuracy)
```
